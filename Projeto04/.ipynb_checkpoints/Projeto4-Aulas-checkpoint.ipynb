> Exercício 1: A coluna "jobedu" parece ter duas informações. Vamos separar em duas colunas.

```python
# Temos uma coluna chamada 'jobedu' que contém duas informações juntas:
# exemplo de valor: "management,tertiary"
# O objetivo é separar essa coluna em duas novas colunas:
# - 'profissao' → primeira parte (antes da vírgula)
# - 'graduacao' → segunda parte (depois da vírgula)

# Para isso, usamos o método .str.split() do pandas:
# - str.split(',') → divide o texto pelo separador vírgula
# - expand=True → cria colunas separadas em vez de listas
# - [['profissao', 'graduacao']] → atribui nomes às novas colunas

# Código:
df[['profissao', 'graduacao']] = df['jobedu'].str.split(',', expand=True)

# Resultado esperado:
# jobedu              | profissao   | graduacao
# --------------------|-------------|-----------
# management,tertiary | management  | tertiary
# blue-collar,primary | blue-collar | primary
# technician,secondary| technician  | secondary

# Se não precisar mais da coluna original 'jobedu', pode removê-la:
df.drop(columns=['jobedu'], inplace=True)
```

## Tratamento da Variavel Age

> Exercício 2: Vamos imputar os valores ausentes da variável age com uma medida de tendência central. Escolha uma das medidas, aplique a imputação e justifique sua escolha. Deixamos a variável como float ou como int? Se convertemos, fazemos isso antes ou depois da imputação?

```python
# Coloque sua soluç# Info
df.info()
```

```python
# Quando usar cada medida
# - Mediana:
# - Melhor em distribuições assimétricas ou com outliers.
# - Representa o ponto central (50% abaixo, 50% acima).
# - No seu caso, como vimos que a idade é assimétrica e há outliers, a mediana tende a ser mais confiável.

# - Moda:
# - Útil quando existe um valor que aparece com muita frequência.
# - Faz sentido se houver uma concentração clara em uma idade específica (ex.: muitos clientes com exatamente 40 anos).
# - Se a distribuição for muito dispersa, a moda pode não representar bem o conjunto.
```

```python
# - Compare com o histograma:
# - Se a moda for um valor isolado (ex.: 40 anos aparece mais, mas não representa o centro da distribuição), prefira a mediana.
# - Se a moda coincidir com o pico principal e fizer sentido como idade típica, pode ser usada.


print("Mediana:", df['age'].median())
print("Moda:", df['age'].mode()[0])
```

```python
# A variável 'age' possui 20 valores ausentes.
# Calculamos:
# - Mediana = 39
# - Moda = 32
#
# Como a distribuição da idade é assimétrica e possui outliers,
# a média não é recomendada. Entre mediana e moda, a mediana
# representa melhor o centro da distribuição.
#
# Portanto, o processo correto é imputar os valores ausentes
# com a mediana (39) e depois converter a coluna para int64:
# df['age'].fillna(df['age'].median(), inplace=True)
# df['age'] = df['age'].astype('int64')
```

```python
# O comando df['age'].fillna(df['age'].median(), inplace=True) faz o seguinte:
#
# 1. df['age'] → seleciona a coluna 'age' do DataFrame.
#
# 2. .fillna(df['age'].median()) → substitui todos os valores ausentes (NaN)
#    pela mediana da própria coluna 'age'. 
#    - A mediana é calculada com df['age'].median(), que no nosso caso é 39.
#    - Isso garante que os 20 valores ausentes sejam preenchidos com 39.
#
# 3. inplace=True → aplica a alteração diretamente no DataFrame original,
#    sem precisar criar uma nova variável ou atribuição.
#
# Em resumo: esse comando imputa os valores ausentes da coluna 'age' com a mediana
# e já atualiza o DataFrame df, deixando a coluna sem valores NaN.

df['age'].fillna(df['age'].median(), inplace=True)
```

```python
# Valores ausentes no dataframe
df.isna().any()
```

```python
# Valores ausentes da variável age
df.age.isnull().sum()

# Calcula o percentual de valores ausentes na variável age
df.age.isnull().mean()*100
```

```python
# A coluna 'age' está no tipo float64, mas como representa idade,
# faz mais sentido convertê-la para int64 (número inteiro).
# Isso evita valores decimais desnecessários e deixa os dados mais consistentes.

# Para converter, usamos o método astype() do pandas:
df['age'] = df['age'].astype('int64')

# Depois da conversão, a coluna 'age' passa a ser do tipo int64.
# Exemplo:
# Antes: 29.0, 35.0, 42.0
# Depois: 29, 35, 42
```

> Exercício 4: Vamos imputar os valores ausentes da variável salary com uma medida de tendência central. Precisamos também tratar os valores iguais a zero. Escolha sua estratégia, aplique a imputação e justifique sua escolha. 

```python
# A variável 'salario' possui 26 registros com valor 0, que não representam salários reais.
# - Histograma: mostra maior concentração de salários na faixa de 60.000 a 80.000.
# - Boxplot: a mediana está em torno de 60.000, com bastante dispersão e alguns outliers.
# - Os 26 registros com valor 0 são problemáticos, pois não representam salários reais e puxam a média para baixo.
# Decisão fundamentada
# - Média (57.008): não é adequada, pois está distorcida pelos valores 0 e pelos outliers.
# - Moda (20.000): não reflete o centro da distribuição, apenas indica o valor mais repetido.
# - Mediana (60.000): é robusta contra outliers e representa melhor o “salário típico” do conjunto.
#➡️ Portanto, a melhor escolha é imputar os valores 0 com a mediana (60.000). Isso mantém a distribuição consistente com o que os gráficos mostram.
```

```python
# Importante entender a diferença entre replace() e fillna():
#
# - df['salary'].replace(0, df['salary'].median(), inplace=True)
#   → Substitui apenas os valores iguais a 0 pela mediana (60.000).
#   → Útil quando temos registros inválidos (salário = 0).
#
# - df['salary'].fillna(df['salary'].median(), inplace=True)
#   → Substitui apenas os valores ausentes (NaN) pela mediana.
#   → Útil quando temos células sem informação de salário.
#
# Ou seja:
# - replace() trata valores existentes que são inválidos (como 0).
# - fillna() trata valores ausentes (NaN).
#
# No caso da coluna 'salary', temos dois problemas diferentes:
# 1. 26 registros com valor 0 → corrigimos com replace().
# 2. 26 valores ausentes (NaN) → corrigimos com fillna().
#
# Portanto, precisamos aplicar os dois comandos para deixar a coluna consistente:
# df['salary'].replace(0, df['salary'].median(), inplace=True)
# df['salary'].fillna(df['salary'].median(), inplace=True)
```

```python
# Decisão: imputar os valores 0 com a mediana (60.000).
df['salary'].replace(0, df['salary'].median(), inplace=True)
df['salary'].fillna(df['salary'].median(), inplace=True)
```

```python
# Valores ausentes na variável
df.salary.isnull().sum()
```

```python
# Verificar valores nulo nas colunas
df.isna().any()
```
